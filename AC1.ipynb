$\frac{dy(t)}{dt} = (y+4) \; cos(t)$

$\frac{1}{y+4}\frac{dy(t)}{dt} - cos(t)=0$

$\frac{1}{y+4}\frac{dy(t)}{dt} - \frac{sin(t)}{dt}=0$

$∫\frac{d(ln(y+4)-sin(t))}{dt}dt = ∫0 + C$

$ln(y+4)-sin(t) = C$

$ln(y+4) = C+sin(t)$

$y(t)= e^{sin(t)+c}-4$

Para $y(0)= 1$, temos que:

$y(0) = e^{sin(0)+c}-4$

$1 = e^{0+c}-4$

$e^c = 5$

$c = ln(5)$

Portanto, nossa $y(t)$ fica :

$y(t) = e^{sin(t)+ln(5)}-4$

É uma função periódica assim como mostrado na plotagem abaixo:

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math

# function that returns dy/dt
def model(y,t):
    k = 4
    dydt = (y+k)*math.cos(t)
    return dydt

# initial condition
y0 = 1

# time points
t = np.linspace(0,20)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

$\frac{dy(t)}{dt} = ye^{t}$

$\frac{1}{y}\frac{dy(t)}{dt} - e^{t} = 0$

$∫\frac{1}{y}\frac{dy(t)}{dt} - \frac{d e^{t}}{dt} dt= \int0dt +c$

$∫\frac{1}{y}\frac{dy(t)}{dt} - \frac{d e^{t}}{dt} dt= \int0dt +c$

$ln(y) -e^t = c$

$ln(y) = e^t + c$

$y(t)= e^{e^t+c}$

como $y(0) =1$, ficamos com :

$1= e^{e^0+c}=e^{e^0+c}=e^{1+c}$

$ln(1)= 1+c$

$c = ln(1) - 1 = -1$

Assim, ficamos com:

$y(t)= e^{e^t-1}$




```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import math

# function that returns dy/dt
def model(y,t):
    dydt = y*math.exp(t)
    return dydt

# initial condition
y0 = 1

# time points
t = np.linspace(0,2)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
